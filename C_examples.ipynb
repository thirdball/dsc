#### This program plays the song "Hot Cross Buns" with an echo playing 0.3 seconds after the start of each note.

##### Question: "What does a song with an echo sound like?"
To answer this, I am adding an 0.3 second delayed echo at 75% volume to the simple song
"Hot Cross Buns".

##### Conclusion:
I think that the jingle sounds pretty good. The echo at the fast part of the song towards the end sounds a bit weird and it could sound better if the volume of the echo was slightly reduced at that part to around 60% or 50%, but I still think that 75% was a good constant volume for the echo. I also that that 0.3 seconds was a good delay for the echo to make it noticeable enough without it being too delayed.
******

```python
# your code
import numpy as np
from IPython.display import Audio

# Generates sound wave with a given note frequency and length
def generate_wave(note, length, sampling_rate=44100):
    # Creates array representing time (t) for the given duration
    t = np.arange(int(sampling_rate * length))
    # Generates the sine wave using the formula sin(2πft)
    wave = np.sin(2 * np.pi * t * note / sampling_rate)
    return wave
    
# Concatenats individual note waves
def generate_melody(notes, lengths, sampling_rate=44100):
    melody = np.concatenate([generate_wave(note, length, sampling_rate) 
                             for note, length in zip(notes, lengths)])
    return melody

# Adds echo effect
def add_echo(signal, delay, attenuation, sampling_rate=44100):
     # Creates an array of zeros with the same shape as the signal
    echo_signal = np.zeros_like(signal)
    # Calculates the number of samples corresponding to the desired delay
    delay_samples = int(sampling_rate * delay)
    
    # For each sample index starting from delay_samples,
    # adds the delayed signal attenuated by the given factor
    for i in range(delay_samples, len(signal)):
        echo_signal[i] = signal[i - delay_samples] * attenuation
    
    return signal + echo_signal

# Note lengths and order (Song: Hot Cross Buns) 
e = 329 
d = 293 
c = 261 

notes = [e, d, c, e, d, c, c, c, c, c, d, d, d, d, e, d, c] 
lengths = [1, 1, 2, 1, 1, 2, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 1, 1, 2]

# Generates the melody
melody = generate_melody(notes, lengths)

# Adds Echo
delay = 0.3  # Delay of echo in seconds
attenuation = 0.75  # Volume of echo
melody_with_echo = add_echo(melody, delay, attenuation)

# Scale the sine wave to fit the WAV standard
scaled_melody_with_echo = np.int16(melody_with_echo / np.max(np.abs(melody_with_echo)) * 32767)

# Display the audio
Audio(scaled_melody_with_echo, rate=44100)
```
